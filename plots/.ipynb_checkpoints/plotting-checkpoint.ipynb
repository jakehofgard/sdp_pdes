```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Relaxation comparison for MOSEK

plt.figure(figsize=(8, 6))
for i in range(1, 6):
    plt.plot(np.loadtxt(f"../outputs/mosek_d={2 * i}.txt")[:, 1], label=fr"$d = {2 * i}$")
plt.legend()
plt.xlabel("Iteration")
plt.ylabel("Optimality Gap")
plt.title(r"Optimality Gap vs. Iteration for Increasing Relaxation Degree $d$, MOSEK")
plt.yscale("log")
plt.savefig("MOSEK_opt_gap_n=2.png", dpi=400)
```

```python
# Relaxation comparison for SeDuMi

sedumi_4 = np.array(pd.read_csv(f"../outputs/sedumi_d=4.txt", sep=" ")["b*y"].dropna())
sedumi_6 = np.array(pd.read_csv(f"../outputs/sedumi_d=6.txt", sep=" ")["Unnamed: 6"].dropna())
sedumi_8 = np.array(pd.read_csv(f"../outputs/sedumi_d=8.txt", sep=" ")["Unnamed: 6"].dropna())
sedumi_10 = np.array(pd.read_csv(f"../outputs/sedumi_d=10.txt", sep=" ")["Unnamed: 6"].dropna())
```

```python
plt.figure(figsize=(8, 6))
plt.plot(sedumi_4, label=r"$d=4$")
plt.plot(sedumi_6, label=r"$d=6$")
plt.plot(sedumi_8, label=r"$d=8$")
plt.plot(sedumi_10, label=r"$d=10$")
plt.yscale("log")
plt.legend()

plt.xlabel("Iteration")
plt.ylabel("Optimality Gap")
plt.title(r"Optimality Gap vs. Iteration for Increasing Relaxation Degree $d$, SeDuMi")
plt.yscale("log")
plt.savefig("SeDuMi_opt_gap_n=2.png", dpi=400)
```

```python
# Curse of dimensionality for SCS

scs_data = pd.read_csv("../outputs/scs_prob_size.txt")
d = np.array(scs_data["d"])
n_vars = np.array(scs_data[" n"])
n_cons = np.array(scs_data[" m"])
time = np.array(scs_data[" t"])
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14, 4))
fig.suptitle('SDP Size and Runtime for SCS')
ax1.plot(d, n_vars)
ax1.set_xlabel(r"$d$")
ax1.set_ylabel(r"Number of Variables")
ax2.plot(d, n_cons)
ax2.set_xlabel(r"$d$")
ax2.set_ylabel(r"Number of Constraints")
ax2.set_yscale("log")
ax3.plot(d, time)
ax3.set_xlabel(r"$d$")
ax3.set_ylabel(r"Runtime (s)")

plt.savefig("SCS_prob_size_n=2.png", dpi=600)
```

```python
# SCS hyperparameter tuning for D-R relaxation parameter alpha

dr_param_data = np.loadtxt("../outputs/hjb_energy_low_dim_d=6.csv", delimiter=",")
alphas = dr_param_data[:35, 0]
bounds = dr_param_data[:35, 1]

plt.figure(figsize=(8, 6))
plt.plot(alphas, bounds, label="SCS Lower Bound")
plt.axhline(0.975592, color="r", linestyle="--", label="Deep BSDE Optimal Control")
plt.xlabel(r"SCS Douglas-Rachford Relaxation Parameter $\alpha$")
plt.ylabel(r"Minimum Energy of Optimal Control $||u / 2||_{2}^2$")
plt.title(r"Minimum Energy Control for Relaxation Degree $d = 6$")
plt.legend()

plt.savefig("hjb_energy_n=2.png", dpi=400)
```

```python
# Plot higher dimensional trials

trials = np.zeros((2, 3))

for d in range(1, 4):
    data_3 = np.loadtxt(f"../outputs/scs_n=3_d={d}.txt", delimiter=" ")
    final_gap_3 = np.abs(data_3[-1, 3])
    data_4 = np.loadtxt(f"../outputs/scs_n=4_d={d}.txt", delimiter=" ")
    final_gap_4 = np.abs(data_4[-1, 4])
    trials[0, d-1] = final_gap_3
    trials[1, d-1] = final_gap_4
```

```python
w = 0.25
fig = plt.subplots(figsize =(8, 6))

b1 = np.arange(3)
b2 = [x + w for x in b1]

plt.bar(b1, trials[0, :], color="cornflowerblue", width = w, label =f'$n=3$')
plt.bar(b2, trials[1, :], color="orange", width = w, label =f'$n=4$')
plt.xlabel(f"Relaxation Degree $d$")
plt.ylabel("Optimality Gap")
plt.title("Optimality Gap for Feasible Control Problem in Higher Dimensions")
plt.xticks([r + w/2 for r in range(3)], ['1', '2', '3'])

plt.yscale("log")
plt.legend()

plt.savefig("hjb_higher_dim.png", dpi=400)
```
